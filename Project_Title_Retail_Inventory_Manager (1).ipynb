**Project Title: Retail Inventory Manager**

**Scenario:**
 You are managing an inventory system for a retail shop. Certain product categories are fixed, but the actual products can be added or removed by the staff.

```python
# Tasks: Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
Product_Categories=("Electronics","Stationery","Clothing","Groceries","Furniture")
print (Product_Categories)
```

```python
# Create a list of products that can be updated dynamically.
Products=("TV","Fridge","Pen","Pencil","Pants","Shirts","Rice","Milk","Chair","Table")
print (Products)
```

```python
# Ask the user to enter a new product and add it to the list.
Products=("TV","Fridge","Pen","Pencil","Pants","Shirts","Rice","Milk","Chair","Table")
new_product_name=input("Enter a new product name: ")
Products = Products + (new_product_name,)
print(Products)
```

```python
# Ask the user to input the index of a product to remove it from the list.
Products=("TV","Fridge","Pen","Pencil","Pants","Shirts","Rice","Milk","Chair","Table","Sugar")

product_index = int(input("Enter index to remove: "))

removed = products.pop(product_index)

print("Removed:", removed)
print("Updated list:", products)
```

```python
# Display the final list of products and the product categories.c\
categories = ("Electronics", "Stationery", "Clothing")
products = ['TV', 'Fridge', 'Pen', 'Pencil', 'Pants', 'Shirts', 'Rice', 'Milk', 'Chair', 'Table']

print("\nFinal Product List:", products)
print("Product Categories:", categories)
```
